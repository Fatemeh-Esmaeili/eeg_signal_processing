# Phase 1: DATA INVENTORY STEPS FOR EEG DATASET

## STEP 1: Directory Structure Mapping

```python
import os
import pandas as pd
from collections import Counter

def map_directory_structure(root_path="PEEG"):
    """Map the complete directory structure"""
    structure = {}
    subject_count = 0
    session_counts = []
    eeg_files = []
    
    for root, dirs, files in os.walk(root_path):
        level = root.replace(root_path, '').count(os.sep)
        indent = ' ' * 2 * level
        print(f"{indent}{os.path.basename(root)}/")
        
        # Count subjects
        if 'sub-' in root:
            subject_count += 1
            
        # Collect EEG files and session info
        for file in files:
            if file.endswith('.set'):
                eeg_files.append(os.path.join(root, file))
                # Extract session info from path
                parts = root.split(os.sep)
                if 'ses-' in root:
                    for part in parts:
                        if 'ses-' in part:
                            session_counts.append(part)
                            break
    
    print(f"\n=== DIRECTORY SUMMARY ===")
    print(f"Total subjects found: {subject_count}")
    print(f"Session distribution: {Counter(session_counts)}")
    print(f"Total EEG files: {len(eeg_files)}")
    
    return eeg_files, subject_count, session_counts

# Run directory mapping
print("Starting Step 1: Directory Structure Mapping...")
eeg_files, subject_count, session_info = map_directory_structure()
```

## STEP 2: Subject and Session Inventory

```python
def create_subject_inventory(eeg_files):
    """Create comprehensive subject-level inventory"""
    inventory = []
    
    for file_path in eeg_files:
        # Parse file path to extract information
        parts = file_path.split(os.sep)
        
        subject_id = next((p for p in parts if 'sub-' in p), 'Unknown')
        session_id = next((p for p in parts if 'ses-' in p), 'Unknown')
        task_type = 'Unknown'
        
        # Extract task type from filename
        filename = os.path.basename(file_path)
        if 'pre' in filename.lower():
            task_type = 'Pre-rehabilitation'
        elif 'post' in filename.lower():
            task_type = 'Post-rehabilitation'
        elif 'run' in filename.lower():
            task_type = 'Task-run'
            
        # Get file size
        file_size_mb = os.path.getsize(file_path) / (1024 * 1024)
        
        inventory.append({
            'subject_id': subject_id,
            'session_id': session_id,
            'task_type': task_type,
            'filename': filename,
            'file_path': file_path,
            'file_size_mb': round(file_size_mb, 2),
            'has_fdt_file': os.path.exists(file_path.replace('.set', '.fdt'))
        })
    
    # Create inventory DataFrame
    df_inventory = pd.DataFrame(inventory)
    
    print("\n=== SUBJECT INVENTORY SUMMARY ===")
    print(f"Total files: {len(df_inventory)}")
    print(f"Unique subjects: {df_inventory['subject_id'].nunique()}")
    print(f"Sessions per subject:")
    print(df_inventory.groupby('subject_id')['session_id'].nunique().describe())
    print(f"\nTask type distribution:")
    print(df_inventory['task_type'].value_counts())
    print(f"\nFile size statistics (MB):")
    print(df_inventory['file_size_mb'].describe())
    
    return df_inventory

# Create subject inventory
df_inventory = create_subject_inventory(eeg_files)
```

```python
def create_subject_inventory(eeg_files):
    """Create comprehensive subject-level inventory"""
    print("Creating detailed subject inventory...")
    inventory = []
    
    for file_path in eeg_files:
        # Parse file path to extract information
        parts = file_path.split(os.sep)
        
        subject_id = next((p for p in parts if 'sub-' in p), 'Unknown')
        session_id = next((p for p in parts if 'ses-' in p), 'Unknown')
        task_type = 'Unknown'
        
        # Extract task type from filename
        filename = os.path.basename(file_path)
        if 'pre' in filename.lower():
            task_type = 'Pre-rehabilitation'
        elif 'post' in filename.lower():
            task_type = 'Post-rehabilitation'
        elif 'run' in filename.lower():
            task_type = 'Task-run'
        elif 'ori' in filename.lower():
            task_type = 'Orientation'
            
        # Get file size
        file_size_mb = os.path.getsize(file_path) / (1024 * 1024)
        
        inventory.append({
            'subject_id': subject_id,
            'session_id': session_id,
            'task_type': task_type,
            'filename': filename,
            'file_path': file_path,
            'file_size_mb': round(file_size_mb, 2),
            'has_fdt_file': os.path.exists(file_path.replace('.set', '.fdt'))
        })
    
    print(f"\nProcessed {len(inventory)} EEG files")
    
    # Basic analysis without pandas
    unique_subjects = set([item['subject_id'] for item in inventory])
    task_types = [item['task_type'] for item in inventory]
    session_counts = [item['session_id'] for item in inventory]
    
    print("\n=== SUBJECT INVENTORY SUMMARY ===")
    print(f"Total files: {len(inventory)}")
    print(f"Unique subjects: {len(unique_subjects)}")
    
    print(f"\nTask type distribution:")
    task_counts = Counter(task_types)
    for task, count in task_counts.items():
        print(f"  {task}: {count} files")
    
    print(f"\nSession distribution:")
    session_counts_dist = Counter(session_counts)
    for session, count in session_counts_dist.items():
        print(f"  {session}: {count} files")
    
    # File sizes
    file_sizes = [item['file_size_mb'] for item in inventory]
    print(f"\nFile size statistics (MB):")
    print(f"  Min: {min(file_sizes):.2f}")
    print(f"  Max: {max(file_sizes):.2f}")
    print(f"  Average: {sum(file_sizes)/len(file_sizes):.2f}")
    
    # Check for missing .fdt files
    missing_fdt = sum(1 for item in inventory if not item['has_fdt_file'])
    print(f"\nMissing .fdt files: {missing_fdt}")
    
    return inventory

# Create subject inventory
print("\n" + "="*50)
print("Step 2: Subject and Session Inventory")
print("="*50)
inventory_data = create_subject_inventory(eeg_files)
```

## Step 4: Data Completeness Analysis

```python
def analyze_data_completeness(inventory_data):
    """Analyze data completeness and identify patterns"""
    print("Analyzing data completeness across subjects and sessions...")
    
    # Group by subject
    subjects_data = {}
    for item in inventory_data:
        subject = item['subject_id']
        if subject not in subjects_data:
            subjects_data[subject] = []
        subjects_data[subject].append(item)
    
    # Analyze each subject
    subject_stats = []
    for subject, files in subjects_data.items():
        sessions = set([f['session_id'] for f in files])
        task_types = set([f['task_type'] for f in files])
        file_count = len(files)
        
        subject_stats.append({
            'subject': subject,
            'sessions': len(sessions),
            'task_types': len(task_types),
            'file_count': file_count,
            'has_pre_post': 'Pre-rehabilitation' in task_types and 'Post-rehabilitation' in task_types
        })
    
    print("\n=== DATA COMPLETENESS ANALYSIS ===")
    print(f"Total subjects: {len(subject_stats)}")
    print(f"Total files: {len(inventory_data)}")
    
    # Session statistics
    sessions_per_subject = [s['sessions'] for s in subject_stats]
    print(f"\nSessions per subject:")
    print(f"  Min: {min(sessions_per_subject)}")
    print(f"  Max: {max(sessions_per_subject)}")
    print(f"  Average: {sum(sessions_per_subject)/len(sessions_per_subject):.1f}")
    
    # Files per subject
    files_per_subject = [s['file_count'] for s in subject_stats]
    print(f"\nFiles per subject:")
    print(f"  Min: {min(files_per_subject)}")
    print(f"  Max: {max(files_per_subject)}")
    print(f"  Average: {sum(files_per_subject)/len(files_per_subject):.1f}")
    
    # Pre/Post analysis
    subjects_with_both = sum(1 for s in subject_stats if s['has_pre_post'])
    print(f"\nSubjects with both Pre and Post rehabilitation data: {subjects_with_both}/{len(subject_stats)}")
    
    # Task type distribution
    task_counts = {}
    for item in inventory_data:
        task = item['task_type']
        task_counts[task] = task_counts.get(task, 0) + 1
    
    print(f"\nDetailed Task Distribution:")
    for task, count in task_counts.items():
        percentage = (count / len(inventory_data)) * 100
        print(f"  {task}: {count} files ({percentage:.1f}%)")
    
    # Session distribution details
    session_counts = {}
    for item in inventory_data:
        session = item['session_id']
        session_counts[session] = session_counts.get(session, 0) + 1
    
    print(f"\nDetailed Session Distribution:")
    for session, count in session_counts.items():
        percentage = (count / len(inventory_data)) * 100
        print(f"  {session}: {count} files ({percentage:.1f}%)")
    
    return subject_stats

# Analyze data completeness
print("\n" + "="*50)
print("Step 4: Data Completeness Analysis")
print("="*50)
subject_stats = analyze_data_completeness(inventory_data)
```

## STEP 5: Metadata Collection Template

```python
def extract_existing_metadata(data_folder="PEEG", output_folder="EEG_Inventory"):
    """Extract and merge existing metadata from JSON files in the data folder"""
    import pandas as pd
    import json
    import os
    import glob
    from collections import defaultdict
    
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    print("ğŸ” Searching for existing metadata JSON files...")
    
    # Find all JSON files in the PEEG folder structure
    json_files = glob.glob(os.path.join(data_folder, "**", "*.json"), recursive=True)
    print(f"Found {len(json_files)} JSON files")
    
    # Dictionary to store metadata by subject
    subject_metadata = defaultdict(dict)
    
    for json_file in json_files:
        try:
            with open(json_file, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            print(f"ğŸ“„ Processing: {os.path.basename(json_file)}")
            print(f"   Raw data: {data}")
            
            # Extract subject ID - handle both numeric and string formats
            subject_id = data.get('subject_id')
            if subject_id is not None:
                # Convert to consistent format (sub-01, sub-02, etc.)
                if isinstance(subject_id, (int, float)):
                    subject_id = f"sub-{int(subject_id):02d}"
                else:
                    subject_id = str(subject_id)
                
                # Map the JSON fields to our standardized format
                mapped_data = {
                    'subject_id': subject_id,
                    'age': data.get('age'),
                    'gender': data.get('sex'),
                    'stroke_type': data.get('diagnosis'),
                    'affected_hemisphere': data.get('affected_side'),  # L/R to Left/Right
                    'time_since_stroke_months': data.get('duration'),
                    'lesion_location': data.get('affected_area'),
                    'height_cm': data.get('height'),
                    'weight_kg': data.get('weight'),
                    'acquisition_date': data.get('acquisition_date'),
                    'paradigm': data.get('paradigm'),
                    'source_file': os.path.basename(json_file)
                }
                
                # Convert affected side to full description
                if mapped_data['affected_hemisphere'] == 'L':
                    mapped_data['affected_hemisphere'] = 'Left'
                elif mapped_data['affected_hemisphere'] == 'R':
                    mapped_data['affected_hemisphere'] = 'Right'
                
                # Convert gender to full description
                if mapped_data['gender'] == 'M':
                    mapped_data['gender'] = 'Male'
                elif mapped_data['gender'] == 'F':
                    mapped_data['gender'] = 'Female'
                
                subject_metadata[subject_id] = mapped_data
                print(f"   âœ… Mapped subject {subject_id}")
                
        except Exception as e:
            print(f"âš ï¸  Error reading {json_file}: {e}")
    
    print(f"\nğŸ“Š Successfully extracted metadata for {len(subject_metadata)} subjects")
    
    # Convert to DataFrame
    if subject_metadata:
        metadata_df = pd.DataFrame(list(subject_metadata.values()))
        
        # Reorder columns for better readability
        column_order = [
            'subject_id', 'age', 'gender', 'height_cm', 'weight_kg',
            'stroke_type', 'affected_hemisphere', 'lesion_location', 
            'time_since_stroke_months', 'acquisition_date', 'paradigm', 'source_file'
        ]
        
        # Only include columns that exist in the data
        existing_columns = [col for col in column_order if col in metadata_df.columns]
        metadata_df = metadata_df[existing_columns]
        
        # Save the extracted metadata
        csv_file = os.path.join(output_folder, "extracted_metadata.csv")
        metadata_df.to_csv(csv_file, index=False)
        
        # Save as JSON for structured data
        json_file = os.path.join(output_folder, "extracted_metadata.json")
        metadata_dict = {
            "metadata_schema": {
                "description": "Extracted clinical metadata from PEEG dataset",
                "extraction_date": pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S'),
                "total_subjects": len(metadata_df),
                "fields": {
                    "subject_id": "Unique subject identifier (formatted as sub-XX)",
                    "age": "Age in years",
                    "gender": "Biological sex",
                    "height_cm": "Height in centimeters", 
                    "weight_kg": "Weight in kilograms",
                    "stroke_type": "Type of stroke diagnosis",
                    "affected_hemisphere": "Brain hemisphere affected by stroke",
                    "lesion_location": "Specific brain region affected",
                    "time_since_stroke_months": "Time since stroke onset in months",
                    "acquisition_date": "Date of data acquisition",
                    "paradigm": "Experimental paradigm",
                    "source_file": "Original JSON file source"
                }
            },
            "subjects": metadata_df.to_dict('records')
        }
        
        with open(json_file, 'w', encoding='utf-8') as f:
            json.dump(metadata_dict, f, indent=2, ensure_ascii=False)
        
        print(f"\n EXTRACTION COMPLETE")
        print(f" CSV file: {csv_file}")
        print(f" JSON file: {json_file}")
        print(f"\nğŸ“‹ Data Summary:")
        print(f"   Subjects: {len(metadata_df)}")
        print(f"   Available fields: {list(metadata_df.columns)}")
        print(f"   Age range: {metadata_df['age'].min()}-{metadata_df['age'].max()}" if 'age' in metadata_df.columns else "   Age: Not available")
        print(f"   Gender distribution: {metadata_df['gender'].value_counts().to_dict()}" if 'gender' in metadata_df.columns else "   Gender: Not available")
        
        return metadata_df
    else:
        print(" No metadata found in JSON files")
        return None

# Extract the existing metadata
print("="*60)
print("EXTRACTING EXISTING METADATA FROM PEEG FOLDER")
print("="*60)
extracted_metadata = extract_existing_metadata()
```

## STEP 6: Data Completeness Analysis

```python
def analyze_data_completeness(df_inventory, folder_name="EEG_Inventory"):
    """Analyze data completeness and identify gaps"""
    import os
    import pandas as pd
    
    # Create folder if it doesn't exist
    os.makedirs(folder_name, exist_ok=True)
    
    # Sessions per subject
    sessions_per_subject = df_inventory.groupby('subject_id')['session_id'].nunique()
    
    # Task types per subject
    tasks_per_subject = df_inventory.groupby('subject_id')['task_type'].nunique()
    
    completeness_report = {
        'total_subjects': df_inventory['subject_id'].nunique(),
        'subjects_with_multiple_sessions': len(sessions_per_subject[sessions_per_subject > 1]),
        'subjects_with_single_session': len(sessions_per_subject[sessions_per_subject == 1]),
        'subjects_with_pre_post_data': len(tasks_per_subject[tasks_per_subject >= 2]),
        'missing_fdt_files': len(df_inventory[~df_inventory['has_fdt_file']]),
        'average_files_per_subject': len(df_inventory) / df_inventory['subject_id'].nunique()
    }
    
    print("\n=== DATA COMPLETENESS REPORT ===")
    for key, value in completeness_report.items():
        print(f"{key}: {value}")
    
    # Identify subjects with incomplete data
    subject_summary = df_inventory.groupby('subject_id').agg({
        'session_id': 'nunique',
        'task_type': 'nunique',
        'file_path': 'count',
        'has_fdt_file': 'all'
    }).rename(columns={
        'session_id': 'n_sessions',
        'task_type': 'n_task_types',
        'file_path': 'n_files',
        'has_fdt_file': 'all_files_complete'
    })
    
    print("\n=== SUBJECT-LEVEL SUMMARY ===")
    print(subject_summary.describe())
    
    # Save files to the folder
    completeness_df = pd.DataFrame([completeness_report])
    completeness_path = os.path.join(folder_name, "data_completeness_report.csv")
    subject_summary_path = os.path.join(folder_name, "subject_level_summary.csv")
    
    completeness_df.to_csv(completeness_path, index=False)
    subject_summary.to_csv(subject_summary_path)
    
    print(f"âœ“ Data completeness report saved to: {completeness_path}")
    print(f"âœ“ Subject level summary saved to: {subject_summary_path}")
    
    return completeness_report, subject_summary

completeness, subject_summary = analyze_data_completeness(df_inventory)
```

### Save as CSV File (Structured Data)

```python
def save_inventory_csv(inventory_data, folder_name="EEG_Inventory", filename="eeg_inventory_data.csv"):
    """Save detailed inventory as CSV file"""
    import csv
    import os
    
    # Create folder if it doesn't exist
    os.makedirs(folder_name, exist_ok=True)
    
    # Create full file path
    full_file_path = os.path.join(folder_name, filename)
    
    with open(full_file_path, 'w', newline='', encoding='utf-8') as csvfile:
        # Include ALL fields from your inventory data
        fieldnames = ['subject_id', 'session_id', 'task_type', 'filename', 'file_path', 'file_size_mb', 'has_fdt_file']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        writer.writeheader()
        for item in inventory_data:
            writer.writerow(item)
    
    print(f"âœ“ Detailed inventory saved as: {full_file_path}")

# Save CSV file
save_inventory_csv(inventory_data)
```

```python
def save_inventory_json(inventory_data, subject_stats, folder_name="EEG_Inventory"):
    """Save inventory as JSON file"""
    import json
    import os
    from datetime import datetime

    # Create main folder
    os.makedirs(folder_name, exist_ok=True)

    # Define the file path inside the folder
    filename = os.path.join(folder_name, "eeg_inventory.json")
    
    inventory_dict = {
        'metadata': {
            'total_subjects': len(subject_stats),
            'total_files': len(inventory_data),
            'creation_date': pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S'),
            'dataset_reference': 'Li et al. (2025) Lower limb motor imagery EEG dataset'
        },
        'summary': {
            'subjects_with_pre_post': sum(1 for s in subject_stats if s['has_pre_post']),
            'average_files_per_subject': sum(s['file_count'] for s in subject_stats)/len(subject_stats),
            'average_sessions_per_subject': sum(s['sessions'] for s in subject_stats)/len(subject_stats)
        },
        'inventory_data': inventory_data,
        'subject_stats': subject_stats
    }
    
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(inventory_dict, f, indent=2, ensure_ascii=False)
    
    print(f"âœ“ JSON inventory saved as: {filename}")

# Save JSON file
save_inventory_json(inventory_data, subject_stats)
```

### Save Multiple Formats

```python
def save_comprehensive_inventory(inventory_data, subject_stats, folder_name="EEG_Inventory", base_filename="eeg_inventory"):
    """Save inventory in multiple formats"""
    import csv
    import os
    
    # Create folder if it doesn't exist
    os.makedirs(folder_name, exist_ok=True)
    
    # 1. Save text report
    def save_text_report():
        report = f"""
EEG STROKE REHABILITATION DATASET - COMPREHENSIVE INVENTORY REPORT
===================================================================

DATASET OVERVIEW:
â€¢ Total Subjects: {len(subject_stats)}
â€¢ Total EEG Files: {len(inventory_data)}
â€¢ Subjects with complete Pre/Post data: {sum(1 for s in subject_stats if s['has_pre_post'])}/{len(subject_stats)} ({sum(1 for s in subject_stats if s['has_pre_post'])/len(subject_stats)*100:.1f}%)
â€¢ Average files per subject: {sum(s['file_count'] for s in subject_stats)/len(subject_stats):.1f}
â€¢ Average sessions per subject: {sum(s['sessions'] for s in subject_stats)/len(subject_stats):.1f}

TASK DISTRIBUTION:
"""
        # Add task distribution
        task_counts = {}
        for item in inventory_data:
            task = item['task_type']
            task_counts[task] = task_counts.get(task, 0) + 1
        
        for task, count in task_counts.items():
            percentage = (count / len(inventory_data)) * 100
            report += f"  {task}: {count} files ({percentage:.1f}%)\n"
        
        report += "\nSESSION DISTRIBUTION:\n"
        
        # Add session distribution
        session_counts = {}
        for item in inventory_data:
            session = item['session_id']
            session_counts[session] = session_counts.get(session, 0) + 1
        
        for session in sorted(session_counts.keys()):
            count = session_counts[session]
            percentage = (count / len(inventory_data)) * 100
            report += f"  {session}: {count} files ({percentage:.1f}%)\n"
        
        # Save to file in folder
        txt_file = os.path.join(folder_name, f"{base_filename}_report.txt")
        with open(txt_file, 'w', encoding='utf-8') as f:
            f.write(report)
        print(f"âœ“ Text report saved as: {txt_file}")
        return txt_file
    
    # 2. Save CSV data
    def save_csv_data():
        csv_file = os.path.join(folder_name, f"{base_filename}_detailed.csv")
        with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['subject_id', 'session_id', 'task_type', 'filename', 'file_path', 'file_size_mb', 'has_fdt_file']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            for item in inventory_data:
                writer.writerow(item)
        print(f"âœ“ Detailed inventory saved as: {csv_file}")
        return csv_file
    
    # 3. Save subject summary as CSV
    def save_subject_summary():
        subject_csv_file = os.path.join(folder_name, f"{base_filename}_subject_summary.csv")
        with open(subject_csv_file, 'w', newline='', encoding='utf-8') as f:
            writer = csv.DictWriter(f, fieldnames=subject_stats[0].keys())
            writer.writeheader()
            for stats in subject_stats:
                writer.writerow(stats)
        print(f"âœ“ Subject summary saved as: {subject_csv_file}")
        return subject_csv_file
    
    # 4. Save session statistics
    def save_session_stats():
        session_stats = {}
        for item in inventory_data:
            session = item['session_id']
            if session not in session_stats:
                session_stats[session] = {'files': 0, 'subjects': set()}
            session_stats[session]['files'] += 1
            session_stats[session]['subjects'].add(item['subject_id'])
        
        session_file = os.path.join(folder_name, f"{base_filename}_session_stats.csv")
        with open(session_file, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['session', 'file_count', 'subject_count'])
            for session, stats in session_stats.items():
                writer.writerow([session, stats['files'], len(stats['subjects'])])
        print(f"âœ“ Session statistics saved as: {session_file}")
        return session_file
    
    # Execute all saving functions
    print("Saving comprehensive inventory...")
    txt_file = save_text_report()
    csv_file = save_csv_data()
    subject_file = save_subject_summary()
    session_file = save_session_stats()
    
    print(f"\nğŸ‰ COMPREHENSIVE INVENTORY SAVED!")
    print(f"   Text Report: {txt_file}")
    print(f"   Detailed Data: {csv_file}")
    print(f"   Subject Summary: {subject_file}")
    print(f"   Session Stats: {session_file}")

# Save everything
save_comprehensive_inventory(inventory_data, subject_stats)
```
